```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
fig, ax = plt.subplots(figsize=(7, 6))

energies = np.genfromtxt("output/energies.out")
dos = np.genfromtxt("output/dos.out")

ax.scatter(energies, dos)
```

```python
fig, ax = plt.subplots(figsize=(7,6))

kpoints = np.genfromtxt("output/kpoints.out")
density = np.genfromtxt("output/density.out")
density_layer = np.genfromtxt("output/density_layer.out")

im = ax.scatter(kpoints[0,:], kpoints[1,:], c=density[2,:], s=0.3, cmap="inferno")
plt.colorbar(im, ax=ax)
ax.set_aspect("equal")
```

```python
fig, ax = plt.subplots(figsize=(7, 6))

colors = density_layer[8, :]

vmax = np.max(np.abs(colors))

im = ax.scatter(kpoints[0, :], kpoints[1, :], vmax=vmax, vmin=-vmax,
                c=colors, s=0.1, cmap="PiYG")
plt.colorbar(im, ax=ax)
ax.set_aspect("equal")
```
