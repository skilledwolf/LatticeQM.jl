```python
import numpy as np
import matplotlib.pyplot as plt

kpoints = np.genfromtxt("output_fs/kpoints.out")
density = np.genfromtxt("output_fs/density.out")
density_layer = np.genfromtxt("output_fs/density_layer.out")

i=5

fig, ax = plt.subplots(figsize=(7, 6))
im = ax.scatter(kpoints[0,:], kpoints[1,:], c=density[i,:], s=0.3, cmap="inferno")
plt.colorbar(im, ax=ax)
ax.set_aspect("equal")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

kpoints = np.genfromtxt("output_fs2/kpoints.out")
density = np.genfromtxt("output_fs2/density.out")
density_layer = np.genfromtxt("output_fs2/density_layer.out")
density_valley = np.genfromtxt("output_fs2/density_valley.out")

i=7

colors = density_valley[i, :]
vmax = np.max(np.abs(colors))

fig, [ax1,ax2] = plt.subplots(1,2, sharey=True, figsize=(12, 5), constrained_layout=True)
im = ax1.scatter(kpoints[0,:], kpoints[1,:], c=density[i,:], s=1, cmap="Greys") #cmap="inferno"
im2 = ax2.scatter(kpoints[0, :], kpoints[1, :], vmax=vmax, vmin=-vmax,
                c=colors, s=0.1, cmap="seismic")

plt.colorbar(im2, ax=ax2)
ax1.set_aspect("equal")
ax2.set_aspect("equal")
# ax.set_xlim(-0.05,0.05)
# ax.set_ylim(-0.05,0.05)
plt.show()
```
